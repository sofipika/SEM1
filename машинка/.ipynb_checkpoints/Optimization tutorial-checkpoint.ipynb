### Optimization

Lets find roots of that square equasion via gradient descent. What is the first thing we will have to do?

```python
def draw_f_plot(x_values, y_values, color):
    plt.rcParams["figure.figsize"] = (10,7)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.grid(True, which='both')
    plt.xticks(np.arange(-5, 10, 1))
    plt.yticks(np.arange(-10, 60,10))
    plt.plot(x,y, color=color)

def f(x):
    return x**2 - 5*x + 4

x = np.linspace(-5,10,100)
y = np.array(list(map(f, x)))


answer_x = np.array([1,4])
answer_y = np.array([0,0])
plt.scatter(answer_x, answer_y, color='r', linewidths=5)
draw_f_plot(x,y, 'b')
```

```python
def draw_f_squared_plot(x_values, y_values, color):
    plt.rcParams["figure.figsize"] = (10,7)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.grid(True, which='both')
    plt.xticks(np.arange(-5, 10, 1))
    plt.yticks(np.arange(-1, 14 ,1))
    plt.ylim(top=14)
    plt.xlim(left=0, right=6)
    plt.plot(x,y, color=color)
    
def square_f(x):
    return (x**2 - 5*x + 4)**2

x = np.linspace(-5,10,100)
y = np.array(list(map(square_f, x))) 
draw_f_squared_plot(x,y, 'b')
answer_x = np.array([1,4])
answer_y = np.array([0,0])
plt.scatter(answer_x, answer_y, color='r', linewidths=5)
```

```python
def square_f(x):
    return (x**2 - 5*x + 4)**2

def grad_square_f(x):
    return 2*(x**2 - 5*x + 4)*(2*x - 5)

def plot_optimum_search(start_point, lr, max_iter):    
    #gradient search
    x_list, y_list = [], []
    curr_x_val = start_point
    for i in range(max_iter):
        clear_output(True)
        prev_x_val = curr_x_val
        prev_y_val = square_f(prev_x_val)
        x_list.append(prev_x_val)
        y_list.append(prev_y_val)
        
        curr_x_val = prev_x_val - lr * grad_square_f(prev_x_val)
        curr_y_val = square_f(curr_x_val)
        x_list.append(curr_x_val)
        y_list.append(curr_y_val)
        
        x_array = np.linspace(-5,10,100)
        y_array = np.array(list(map(square_f, x_array))) 
        draw_f_squared_plot(x_array, y_array, 'b')

        plt.plot(x_list, y_list, 'ro--', linewidth=2, markersize=7)
        plt.show()
        time.sleep(1)
    print('Optimum x: ', curr_x_val)
    

plot_optimum_search(5, 0.05, 10)
```

Let's note - we can stuck in local optimum if get wrong initial start point.

#### Solve this equasion using MSE loss

```python
def square_f(x):
    return (x**2 - 5*x + 4)**2

def grad_square_f(x):
    return 2*(x**2 - 5*x + 4)*(2*x - 5)

def find_by_loss(start_point, lr, max_iter):
    """YOUR CODE HERE"""
    x = start_point
    i = 0
    while not (square_f(x)<0.0001) and i <= max_iter:
        x -= lr * grad_square_f(x)
        i+=1
    return x
find_by_loss(5, 0.01, 1000)
```

Now lets move to 3 dimension plane

```python
def f(x):
    x1, x2 = list(x)
    return (x1**2 + x2 - 2)**2 + (x1 + x2**2 - 15)**2

def plot_f_2_dim():
    plt.rcParams["figure.figsize"] = (10,7)
    X1, X2 = np.meshgrid(np.linspace(-6, 6, 50), np.linspace(-6, 6, 50))
    Z = f([X1, X2])
    levels = list(1.7 ** np.linspace(0, 10, 30) - 1.) + [300]
    plt.contourf(X1, X2, np.sqrt(Z), levels=np.sqrt(levels), cmap=plt.cm.Oranges_r)
    plt.xlim((-5,5))
    plt.ylim((-6,6))
    
    
plot_f_2_dim()
answer_x1 = np.array([-2.486,2.357])
answer_x2 = np.array([-4.181,-3.555])
plt.scatter(answer_x1, answer_x2, color='g', linewidths=5)
```

```python
def f_grad(x):
    x1, x2 = list(x)
    df_x1 = 2 * (x1**2 + x2 - 2)*(2*x1) + 2*(x1 + x2**2 - 15)
    df_x2 = 2 * (x1**2 + x2 - 2) + 2*(x1 + x2**2 - 15) * 2*x2
    return np.array([df_x1, df_x2])


def plot_gd_vs_momentum(start_point, lr, max_iter):    
    grad_x_list  = [start_point]
    momentum_x_list = [start_point]
    
    grad_current_x = start_point
    momentum_current_x = start_point
    curr_v = 0
    beta = 0.5
    for i in range(max_iter):
        clear_output(True)
        grad_prev_x = grad_current_x
        grad_current_x = grad_prev_x - lr * f_grad(grad_prev_x)
        
        grad_x_list.append(grad_current_x)
        
        prev_v = curr_v
        momentum_prev_x = momentum_current_x
        curr_v = beta*prev_v + lr*f_grad(momentum_prev_x)
        momentum_current_x = momentum_prev_x - curr_v
        prev_v = curr_v
        
        momentum_x_list.append(momentum_current_x)
        
        plot_f_2_dim()
        #plot gradient descent
        grad_x_array = np.array(grad_x_list)
        plt.plot(grad_x_array[:, 0], grad_x_array[:, 1], 'ro--', linewidth=2, markersize=10, alpha=0.5, label='GD')
        
        #plot grad with momentum
        momentum_x_array = np.array(momentum_x_list)
        plt.plot(momentum_x_array[:, 0], momentum_x_array[:, 1], 'bo--', linewidth=2, markersize=7, label='momentum')
        plt.legend()
        plt.show()
        time.sleep(1)
    print('Optimum gd x: ', grad_current_x)
    print('Optimum gd with momentum x: ', momentum_current_x)
    
x_0 = np.array([0,0])
lr = 0.01
max_iter = 10

plot_gd_vs_momentum(x_0, lr, max_iter)
```
