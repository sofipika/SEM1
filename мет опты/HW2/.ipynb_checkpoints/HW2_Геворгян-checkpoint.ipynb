```python
import numpy as np
import random
import time
import math
import matplotlib.pyplot as plt
```

```python
np.random.seed(1) 
size = 15
A = np.random.randn(size, size) 
A = A.T @ A + 1e-1 * np.eye(size)
b = np.random.randn(size) 

L=np.linalg.eigvals(A).max()   
mu=np.linalg.eigvals(A).min()

def my_f(x):
    return 1/2 * np.transpose(x) @ A @ x + np.transpose(b) @ x

def my_gradf(x):
    return A @ np.transpose(x) + np.transpose(b)
```

```python
def gradient_descent(eps,g,grad_g,x_0,L_0):
    start_time = time.time()
    x = np.array([0]*size)
    x_prev = x_0.copy()
    alpha = 1/L_0
    k = 0
    step,history = [],[]
    while np.linalg.norm(grad_g(x)) >= eps:
        x_prev, x = x, x_prev - alpha * grad_g(x_prev)
        step.append(k)
        history.append(np.linalg.norm(grad_g(x_prev)))
        k += 1
        if k >= iteration_limit:
            break
    return x
```

 ### 1. Наискорейший градиентный спуск

```python
def gradient_speed(eps):
    start_time = time.time()
    k = 0
    step = []
    history = []
    x_prev,x = np.array([0]*size),np.array([0]*size) 
    while np.linalg.norm(my_gradf(x)) >= eps:
        alpha = (np.linalg.norm(my_gradf(x_prev)) ** 2)/(my_gradf(x_prev) @ A @ np.transpose(my_gradf(x_prev)))
        x_prev, x = x ,x_prev - alpha * my_gradf(x_prev)
        history.append(np.linalg.norm(my_gradf(x_prev)))
        step.append(k)
        k += 1
        if k >= iteration_limit:
            break
    print(f"Minimum value for method gradient_speed: {str(my_f(x))}")
    return (step,history,time.time()-start_time)
```

### 2. Адаптивный градиентный спуск

```python
def gradient_adaptive(eps,g,grad_g,x_0,L_f):
    start_time = time.time()
    L_0 = L_f
    k = 0
    step = []
    history = []
    x = np.array([0]*size)
    x_prev = x_0.copy()
    while np.linalg.norm(grad_g(x))>=eps:
        L_0 /= 2
        x = x_prev - (1/L_0) * grad_g(x_prev)
        if g(x) <= g(x_prev) + grad_g(x_prev) @ np.transpose(x - x_prev) + (L_0/2) * (np.linalg.norm(x - x_prev)) **2:
            x_prev = x
        else:
            L_0 *= 4
        history.append(np.linalg.norm(grad_g(x_prev)))
        step.append(k)
        k += 1
        if k >= iteration_limit:
            break
    print(f"Minimum value for method gradient_adaptive: {str(g(x_prev))}")
    return (step,history,time.time()-start_time)
```

### Ускоренный метод Нестерова

```python
class Nesterov:
    def __init__(self, alpha=1/L, name='Nesterov'):
        self.name = name
        self.alpha = alpha
        self.history = []
        self.step = []
    
    def solve(self, x0, f, gradf, tol=1e-3, N=10000):
        start_time = time.time()
        alpha = self.alpha
        k = 0
        x = np.array([0] * size)
        y = np.array([0] * size)
        x_prev = x0.copy()
        while k == 0 or np.linalg.norm(gradf(x)) > tol: 
            x_prev, x = x, y - alpha * gradf(y)
            y = x + (k/(k+3)) * (x - x_prev)
            self.history.append(np.linalg.norm(gradf(x_prev)))
            self.step.append(k)
            k += 1
            if k >= N: 
                break
        print(f"Minimum value for method nesterov: {str(f(x))}")
        return (self.step,self.history,time.time()-start_time)
```

### Метод тяжелого шарика

```python
class HeavyBall:
    def __init__(self, alpha=None, beta=None, L=L, mu=mu, name='HeavyBall'):
        self.name = name
        if alpha is not None and beta is not None:
            self.alpha = alpha
            self.beta = beta
        else:
            self.alpha, self.beta = self.get_alpha_beta(L, mu)
        self.history = []
        self.step = []
    
    @staticmethod
    def get_alpha_beta(L, mu):
        alpha = 4 / (np.sqrt(L) + np.sqrt(mu))**2
        beta1 = (1 - np.sqrt(alpha * L))**2
        beta2 = (1 - np.sqrt(alpha * mu))**2
        beta = max(beta1, beta2)
        return alpha, beta
        
    def solve(self, x0, f, gradf, tol=0.001, N=10000):
        start_time = time.time()
        alpha, beta = self.alpha, self.beta
        x_prev = np.array([0]*size)
        k = 0
        x = x_prev-(1/L)*gradf(x_prev)
        while k == 0 or np.linalg.norm(gradf(x_prev)) >= tol: 
            x_prev, x = x, x - alpha * gradf(x) + beta * (x - x_prev)
            self.history.append(np.linalg.norm(gradf(x_prev)))
            self.step.append(k)
            if k >= N:
                break
            k += 1
        print(f"Minimum value for method heavyball: {str(f(x))}")
        return (self.step,self.history,time.time()-start_time)
```

### 5. Реализуйте устойчивый подсчет функции и градиента

```python
def LogSumExp(x):
    x_max = x.max()
    return x_max+math.log(sum([math.e ** (A[i] @ np.transpose(x - x_max) + b[i]) for i in range(len(A))])) + 0.5 * np.linalg.norm(x) ** 2

def grad_LogSumExp(x):
    x_max = x.max()
    return  (x+(sum([A[i]*math.e**(A[i]@np.transpose(x-x_max)+b[i]) for i in range(len(A))])/sum([math.e**(A[i]@np.transpose(x-x_max)+b[i]) for i in range(len(A))])))
```
